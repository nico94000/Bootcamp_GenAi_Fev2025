```python
import matplotlib.pyplot as plt
import seaborn as sns

# Création du scatter plot avec les bons noms de colonnes
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x="Social support", y="Score", color="blue")

# Ajouter des titres et labels
plt.xlabel("Soutien Social")
plt.ylabel("Score de Bonheur")
plt.title("Relation entre le Soutien Social et le Score de Bonheur")

# Afficher le graphique
plt.show()
```

```python
# Grouper par région et prendre la moyenne des indicateurs
df_region = df.groupby("Country or region")[["GDP per capita", "Healthy life expectancy"]].mean().reset_index()

# Création des axes
fig, ax1 = plt.subplots(figsize=(10, 5))

# Graphique à barres pour le PIB par habitant
ax1.bar(df_region["Country or region"], df_region["GDP per capita"], color="skyblue", label="PIB par Habitant")
ax1.set_xlabel("Pays")
ax1.set_ylabel("PIB par Habitant ($)", color="blue")
ax1.tick_params(axis="y", labelcolor="blue")
ax1.set_xticklabels(df_region["Country or region"], rotation=90)

# Ajouter un second axe pour l'espérance de vie
ax2 = ax1.twinx()
ax2.plot(df_region["Country or region"], df_region["Healthy life expectancy"], color="red", marker="o", label="Espérance de Vie")
ax2.set_ylabel("Espérance de Vie (années)", color="red")
ax2.tick_params(axis="y", labelcolor="red")

# Ajouter un titre et afficher le graphique
plt.title("Comparaison entre le PIB et l'Espérance de Vie par Pays")
fig.tight_layout()
plt.show()
```
