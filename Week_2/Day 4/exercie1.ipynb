```python
import matplotlib.pyplot as plt
import numpy as np

# Définition des valeurs de x et y
x = np.arange(0, 11, 1)  # Utilisation de np.arange() au lieu de np.range()
y = x ** 2  # y = x^2

# Création du graphique
plt.plot(x, y, marker='o', linestyle='-', color='b', label='y = x^2')  # Tracé du graphe
plt.xlabel("X values")  # Label de l'axe X
plt.ylabel("Y values")  # Label de l'axe Y
plt.title("Line Plot: y = x^2")  # Titre du graphique
plt.legend()  # Ajout de la légende
plt.grid(True)  # Ajout d'une grille
plt.show()  # Affichage du graphique
```

```python
import matplotlib.pyplot as plt

# Données des ventes
products = ["Product A", "Product B", "Product C", "Product D", "Product E"]
sales = [50, 80, 30, 70, 90]  # Nombre d'unités vendues

# Création du graphique en barres
plt.figure(figsize=(6, 4))  # Taille de la figure
plt.bar(products, sales, color=['red', 'blue', 'green', 'purple', 'orange'], edgecolor='black')

# Ajout des labels et du titre
plt.xlabel("Products")
plt.ylabel("Sales")
plt.title("Sales of 5 Products")

# Affichage du graphique
plt.show()
```
