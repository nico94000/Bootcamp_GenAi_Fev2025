```python
model = keras.Sequential([
    layers.Conv2D(32, (3,3), activation='relu', input_shape=(32,32,3)),
    layers.MaxPooling2D((2,2)),
    layers.Conv2D(64, (3,3), activation='relu'),
    layers.MaxPooling2D((2,2)),
    layers.Conv2D(128, (3,3), activation='relu'),
    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dense(10, activation='softmax')
])
```

```python
# Charger et prétraiter le dataset CIFAR-10
from tensorflow.keras.datasets import cifar10
(x_train, y_train), (x_test, y_test) = cifar10.load_data()

# Normalisation des pixels entre 0 et 1
x_train, x_test = x_train / 255.0, x_test / 255.0

# Conversion des labels en one-hot encoding
from tensorflow.keras.utils import to_categorical
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)
```

```python
# Compiler le modèle
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
history = model.fit(x_train, y_train, epochs=10, batch_size=64, validation_data=(x_test, y_test), verbose=1)
```

```python
# Évaluer le modèle
test_loss, test_acc = model.evaluate(x_test, y_test, verbose=1)
print(f"Test Accuracy: {test_acc:.4f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# 1️⃣ Génération des données
np.random.seed(0)
x = np.arange(-1, 1, 0.1).reshape(-1, 1)  # Reshape pour scikit-learn
y = -x**2 + np.random.normal(0, 0.05, len(x))

# 2️⃣ Visualisation des données
plt.figure(figsize=(6,4))
plt.scatter(x, y, color='blue', label='Données avec bruit')
plt.xlabel("x")
plt.ylabel("y")
plt.title("Données générées avec du bruit")
plt.legend()
plt.show()

# 3️⃣ Division des données en train/test
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.4, random_state=0)

# 4️⃣ Entraînement du modèle de régression linéaire
model = LinearRegression()
model.fit(X_train, y_train)

# 5️⃣ Prédictions et évaluation
y_pred = model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)

# 6️⃣ Affichage des résultats
plt.figure(figsize=(6,4))
plt.scatter(X_test, y_test, color='blue', label='Données réelles')
plt.plot(X_test, y_pred, color='red', linewidth=2, label='Prédiction')
plt.xlabel("x")
plt.ylabel("y")
plt.title(f"Régression linéaire - MSE: {mse:.4f}")
plt.legend()
plt.show()

# 7️⃣ Affichage de l'erreur
print(f"Erreur quadratique moyenne (MSE) : {mse:.4f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline
from sklearn.metrics import mean_squared_error

# 1️⃣ Génération des données avec bruit
np.random.seed(0)
x = np.arange(-1, 1, 0.1).reshape(-1, 1)  # Reshape pour scikit-learn
y = -x**2 + np.random.normal(0, 0.05, len(x))

# Division en train/test
X_train, X_test, y_train, y_test = x[:12], x[12:], y[:12], y[12:]

# 2️⃣ Fonction pour ajuster un modèle polynomial
def polynomial_fit(degree):
    """Crée un modèle polynomial de degré 'degree' et l'entraîne."""
    model = make_pipeline(PolynomialFeatures(degree), LinearRegression())
    model.fit(X_train, y_train)
    return model

# 3️⃣ Fonction pour tracer les résultats
def plot_polyfit(degree):
    """Affiche le modèle polynomial ajusté sur les données."""
    model = polynomial_fit(degree)
    
    # Prédiction sur un ensemble dense pour visualisation
    x_dense = np.linspace(-1, 1, 100).reshape(-1, 1)
    y_pred = model.predict(x_dense)
    
    # Prédiction sur l'ensemble de test
    y_test_pred = model.predict(X_test)
    
    # Calcul de l'erreur quadratique moyenne
    mse = mean_squared_error(y_test, y_test_pred)
    
    # Affichage
    plt.figure(figsize=(6,4))
    plt.scatter(X_train, y_train, color='blue', label='Ensemble d\'entraînement')
    plt.scatter(X_test, y_test, color='green', label='Ensemble de test')
    plt.plot(x_dense, y_pred, color='red', linewidth=2, label=f'Polynôme degré {degree}')
    plt.xlabel("x")
    plt.ylabel("y")
    plt.title(f"Régression polynomiale (degré {degree}) - MSE: {mse:.4f}")
    plt.legend()
    plt.show()
    
    print(f"Degré {degree} - Erreur quadratique moyenne (MSE) : {mse:.4f}")

# 4️⃣ Visualisation pour les degrés 1, 7 et 11
for d in [1, 7, 11]:
    plot_polyfit(d)
```
