```python
import requests
from bs4 import BeautifulSoup

# URL de l'annonce
url = "https://www.seloger.com/classified-search?distributionTypes=Buy,Buy_Auction,Compulsory_Auction&estateTypes=House,Apartment&locations=AD08FR36687&order=Default&m=homepage_new_search_classified_search_result"  # Mettre l'URL exacte

# Simuler un navigateur pour éviter d'être bloqué
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/110.0.0.0 Safari/537.36"
}

# Récupérer le contenu HTML
response = requests.get(url, headers=headers)

# Vérifier si l'accès est autorisé
if response.status_code == 200:
    html_content = response.text
    print("✅ Page récupérée avec succès !")
else:
    print("❌ Accès refusé. Vérifie si le site bloque le scraping.")
```

```python

from bs4 import BeautifulSoup

# Parser le HTML
soup = BeautifulSoup(html_content, "html.parser")

# Afficher le HTML formaté
print(soup.prettify()[:2000])  # Afficher les 2000 premiers caractères du HTML
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time

# Configurer Selenium avec Chrome
options = webdriver.ChromeOptions()
options.add_argument("--headless")  # Mode invisible
options.add_argument("--no-sandbox")
options.add_argument("--disable-dev-shm-usage")

# Lancer le navigateur
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)

# Ouvrir la page
url = "https://www.seloger.com/classified-search?distributionTypes=Buy&estateTypes=House,Apartment&locationIds=94028"
driver.get(url)

# Attendre que la page charge
time.sleep(5)

print("✅ Page chargée avec Selenium !")
```

```python
# Trouver les annonces
annonces = driver.find_elements(By.CLASS_NAME, "ListAds__Item")

# Parcourir les annonces et extraire les infos
for annonce in annonces[:5]:  # Limite aux 5 premières
    try:
        titre = annonce.find_element(By.TAG_NAME, "h2").text
        prix = annonce.find_element(By.CLASS_NAME, "Price__Amount").text
        surface = annonce.find_element(By.CLASS_NAME, "Content__PropertySize").text
        
        print(f"🏡 Titre : {titre}")
        print(f"💰 Prix : {prix}")
        print(f"📏 Surface : {surface}")
        print("-" * 40)

    except Exception as e:
        print("⚠️ Problème avec une annonce :", e)

# Fermer le navigateur
driver.quit()
```
