```python

plt.figure(figsize=(8, 5))
sns.countplot(x=df["price_range"], palette="coolwarm")
plt.title("RÃ©partition des CatÃ©gories de Prix")
plt.xlabel("CatÃ©gorie de Prix (0 = Bas, 3 = Ã‰levÃ©)")
plt.ylabel("Nombre de TÃ©lÃ©phones")
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Heatmap des CorrÃ©lations entre les Variables")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x=df["price_range"], y=df["ram"], palette="coolwarm")
plt.title("Impact de la RAM sur le Prix")
plt.xlabel("CatÃ©gorie de Prix")
plt.ylabel("RAM (Mo)")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x=df["price_range"], y=df["battery_power"], palette="coolwarm")
plt.title("Impact de la Batterie sur le Prix")
plt.xlabel("CatÃ©gorie de Prix")
plt.ylabel("CapacitÃ© de la Batterie (mAh)")
plt.show()
```
