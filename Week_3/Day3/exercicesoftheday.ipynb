```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
mean = 50
std = 10
```

```python
mu = 50  # Moyenne
sigma = 10 
```

```python
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
y = norm.pdf(x, mu, sigma)
```

```python
plt.figure(figsize=(8, 5))
plt.plot(x, y, label=f'Normale(μ={mu}, σ={sigma})', color='blue')
plt.title("Distribution Normale")
plt.xlabel("Valeurs")
plt.ylabel("Densité de probabilité")
plt.legend()
plt.grid()
```

exercice 5

```python
import numpy as np
from scipy import stats

# Example data
house_sizes = np.array([50, 70, 80, 100, 120])
house_prices = np.array([150000, 200000, 210000, 250000, 280000])

# Performing linear regression
slope, intercept, r_value, p_value, std_err = stats.linregress(house_sizes, house_prices)

# The slope and intercept of the regression line
print("Slope:", slope)
print("Intercept:", intercept)
```

```python
predicted_price_90 = slope * 90 + intercept
```

```python
predicted_price_90
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(house_sizes, house_prices, color='blue', label='Données réelles')  # Points de données
plt.plot(house_sizes, slope * house_sizes + intercept, color='red', label='Régression linéaire')  # Ligne de régression
```

```python
plt.scatter(90, predicted_price_90, color='green', marker='o', label='Prévision pour 90m²')
```

```python
plt.xlabel("Surface du logement (m²)")
plt.ylabel("Prix du logement (€)")
plt.title("Régression linéaire entre la surface et le prix des logements")
plt.legend()
plt.grid()
```

Nous pouvond donc interpreter qu'il ya un lien entre les m2 et le prix .

exercice 6

```python
import numpy as np
from scipy import stats
```

```python
fertilizer_1 = [5, 6, 7, 6, 5]
fertilizer_2 = [7, 8, 7, 9, 8]
fertilizer_3 = [4, 5, 4, 3, 4]
```

```python

f_statistic, p_value = stats.f_oneway(fertilizer_1, fertilizer_2, fertilizer_3)

```

```python
print("F-value:", f_statistic)
print("P-value:", p_value)
```

```python
alpha = 0.05  
```

```python
if p_value < alpha:
    print("\nLes engrais ont des effets significativement différents sur la croissance des plantes (p < 0.05).")
else:
    print("\nIl n'y a pas de preuve suffisante pour conclure que les engrais ont des effets différents (p > 0.05).")
```

```python
if p_value > 0.05:
    print("\nSi la valeur P était supérieure à 0.05, nous ne pourrions pas rejeter l'hypothèse nulle, ce qui signifie que "
          "les différences observées dans la croissance des plantes pourraient être dues au hasard plutôt qu'à l'effet des engrais.")
```
